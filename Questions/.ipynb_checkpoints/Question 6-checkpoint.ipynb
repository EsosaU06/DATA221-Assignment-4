```python
from tensorflow.keras.datasets import fashion_mnist
from keras import models, layers
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
# entire dataset is divided by 255, the colour values are now normalized

# add channel dimension
X_train = X_train[..., None]
X_test = X_test[..., None]
```

```python
print(X_train.shape)
```

```python
fashionCNNModel = models.Sequential([
    layers.Conv2D(filters=16, kernel_size=(3,3), padding="same", activation="relu"),
    layers.MaxPool2D(),
    layers.Conv2D(filters=32, kernel_size=(3,3), padding="same", activation="relu"),
    layers.MaxPool2D(),
    layers.Flatten(),
    layers.Dense(units=10, activation="softmax") # 10 neurons for 10 possible outputs
])
```

```python
fashionCNNModel.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
fashionCNNModel.fit(X_train, y_train, epochs=15)
```

```python
fashionCNNModel_testAccuracy = fashionCNNModel.evaluate(X_test, y_test)
print(fashionCNNModel_testAccuracy) # 0.9106 accuracy
```
