```python
import pandas
from sklearn.datasets import load_breast_cancer
breastCancerData = load_breast_cancer(as_frame=True) # turning it into a pandas dataframe
```

```python
breastCancerData_X = breastCancerData.data
print(breastCancerData_X.shape) # The X matrix has 569 rows and 30 columns, this is the shape of x
```

```python
breastCancerData_y = breastCancerData.target
print(breastCancerData_y.shape)  # The target has a shape of 569, as there are 569 samples in the vector
```

```python
positive = 0
negative = 0
for i in breastCancerData_y:
    if i == 1:
        positive +=1
    else:
        negative +=1

print("has breast cancer: ",positive) # 357
print("no breast cancer: ",negative) # 212
print("total: ",positive + negative) #569


"""
Since there is not a major difference between the positive and negative target
values (around 60% positive and 40% negative), the dataset is not imbalanced.

Class balance is an important consideration because classification models
learn more from majority classes, meaning it will be more inclined
to guess every sample as belonging to the class belonging to the majority
of the samples in its training dataset.
"""
```
